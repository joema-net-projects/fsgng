This is meant to get some good feng shui numbers that matches the zodiacs of rabbit and dragon.
Other than that, it is no more than a glorified number generator.

The only user addressable things are maxLength and maxIterations.
maxLength is the length of the number set you want (i.e. 4 digits total, you put 4)
maxIterations is how many numbers to generate (i.e. 100 will output 100 4 digit numbers per maxLength)

** 中文翻译 - 百分85是用Google **

这是为了得到一些与兔子和龙的生肖相配的好风水的数字。
除此之外，它只不过是一个数字生成器。
您唯一可以选择的东西是 maxLength 和 maxIterations。
maxLength 是您想要的数字集的长度（例如 即总共 4 位，您输入 4）
maxIterations 是要生产多少遍（例如 100 会输出 100 个 4 位数字)

```python
import random
#to get good feng shui numbers
#Change only maxLength and maxIterations
maxLength = 13 #length of array to get
maxIterations = 100 #Total runs to do

valid = [1,2,3,4,5,6,7,8,9]
ending = [0,1,2,5,6,7]
number = [1,3,5,7,8,9,13,15,17,23,24,31,33,35,37,39,41,48]

iterations = 0
while(iterations < maxIterations):
    someRandomThing = []
    count = 0
    for x in range(maxLength):
        if x == maxLength - 1:
            someRandomThing.append(ending[random.randrange(0, len(ending))])
        else:
            someRandomThing.append(valid[random.randrange(0, len(valid))])
        count += someRandomThing[x]

    if count in number:
        print("Count(总数):",count)
        print(someRandomThing)
        print('')
        iterations += 1 #Successful role
```
